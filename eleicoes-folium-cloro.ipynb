```python
import folium
```

```python
rj = folium.Map(
    location=[-22.65, -43.2]
    ,zoom_start=12
)
```

```python
import json
from pprint import pprint

bairros = 'dc.json'
with open(bairros) as f:
    data = json.load(f)
#pprint(data)
```

```python
folium.GeoJson(
    data,
    name='geojson'
).add_to(rj)
```

```python
import json
from pprint import pprint

bairros = 'bairros_dc.geojson'
with open(bairros) as f:
    data = json.load(f)
#pprint(data)
```

```python
# colocando os nomes dos bairros em maisculo para dar match com o dataframe
for x in data['features']:
    x['properties']['Name'] = x['properties']['Name'].upper()
bairros = [x['properties']['Name'] for x in data['features']]
```

```python
folium.GeoJson(
    data,
    name='geojson'
).add_to(rj)
```

```python
rj.choropleth(geo_data=data
              , data=df
              , columns=['bairro', 'votos_nulos']
              , key_on='properties.Name'
              , fill_color='YlGn'
              , fill_opacity=0.7
              , line_opacity=0.2
              , legend_name='votos nulos')
```
